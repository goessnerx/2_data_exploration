# Context

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.set_theme(style="whitegrid")
%matplotlib inline

trainData = pd.read_csv('train.csv')
trainData
```

# Data quality assessment

# Количество пропущенных значений в разрезе столбцов

```python
nullData = trainData.isnull().sum(axis=0).reset_index().rename(columns={"index": "column", 0: "countNAN"})
fig, ax = plt.subplots(figsize=(12,5))
plt.xticks(rotation=45)
sns.barplot(x="column", y="countNAN", data=nullData)
```

# Наличие дублей

```python
trainWithoutId = trainData.loc[:,'Name':'Embarked']
trainWithoutId[trainWithoutId.duplicated()]
```

# Data exploration

# Соотношение выживших и умерших пассажиров. Более половины пассажиров погибли

```python
arrSurvDie = [trainData[trainData['Survived'] == 1].shape[0] / len(trainData), \
              trainData[trainData['Survived'] == 0].shape[0] / len(trainData)]
labels = ['% выживших', '% погибших']
fig, ax = plt.subplots()
ax.pie(arrSurvDie, labels=labels, wedgeprops=dict(width=0.5), autopct='%1.1f%%', radius=1.5)
```

# Соотношение числа пассажирова в разрезе класса. 3й класс самый многочисленный

```python
dataPclass = trainData.groupby('Pclass').size().reset_index(name='Count').sort_values(by='Pclass')
dataPclass['Percent'] = (dataPclass['Count'] / len(trainData)) * 100
arrPercentPclass = dataPclass.drop(columns = ['Count'],axis = 1)['Percent'].tolist()
labels = ['1 класс', '2 класс', '3 класс']
fig, ax = plt.subplots()
ax.pie(arrPercentPclass, labels=labels, wedgeprops=dict(width=0.5), autopct='%1.1f%%', radius=1.5)
```

# Сравнение шансов на выживание в зависимости от класса пассажира. Как видно, пассажиры первого класса имели больше шансов

```python
dataClassSurvDie = trainData[trainData['Survived'] == 1][['Survived', 'Pclass']].groupby(['Pclass']).size().reset_index(name='countSurv')
dataClassSurvDie = dataClassSurvDie.merge(trainData[trainData['Survived'] == 0][['Survived', 'Pclass']].groupby(['Pclass']).size().reset_index(name='countDie') \
                                         , left_on='Pclass', right_on='Pclass')

dataClassSurvDie['countSurv'] = (dataClassSurvDie[['countSurv']] / len(trainData)) * 100
dataClassSurvDie['countDie'] = (dataClassSurvDie[['countDie']] / len(trainData)) * 100
dataClassSurvDie = dataClassSurvDie.rename(columns={"countSurv": "percentSurv", 'countDie': "percentDie"})
fig, ax = plt.subplots()
dataClassSurvDie.plot(kind='bar', ax=ax, stacked=True, x="Pclass", xlabel='Класс пассажира', ylabel='%', figsize=(12,5))
plt.legend(["Выжившие", "Погибшие"])
```

# Соотношение пассажирова мужского и женского полов. Примерно ранозначное распределение

```python
arrManWoman = [trainData[trainData['Sex'] == 'male'].shape[0] / len(trainData), \
              trainData[trainData['Sex'] == 'female'].shape[0] / len(trainData)]
labels = ['% мужчин', '% женщин']
fig, ax = plt.subplots()
ax.pie(arrManWoman, labels=labels, wedgeprops=dict(width=0.5), autopct='%1.1f%%', radius=1.5)
```

# Сравнение шансов на выживание в зависимости от пола. Женщин выжило значительно больше

```python
dataSexSurvDie = trainData[trainData['Survived'] == 1][['Survived', 'Sex']].groupby(['Sex']).size().reset_index(name='countSurv')
dataSexSurvDie = dataSexSurvDie.merge(trainData[trainData['Survived'] == 0][['Survived', 'Sex']].groupby(['Sex']).size().reset_index(name='countDie') \
                                         , left_on='Sex', right_on='Sex')

dataSexSurvDie['countSurv'] = (dataSexSurvDie[['countSurv']] / len(trainData)) * 100
dataSexSurvDie['countDie'] = (dataSexSurvDie[['countDie']] / len(trainData)) * 100
dataSexSurvDie = dataSexSurvDie.rename(columns={"countSurv": "percentSurv", 'countDie': "percentDie"})
fig, ax = plt.subplots()
dataSexSurvDie.plot(kind='bar', ax=ax, stacked=True, x="Sex", xlabel='Пол пассажира', ylabel='%', figsize=(12,5))
plt.legend(["Выжившие", "Погибшие"])
```

# Анализ влияния возрастной группы на вероятность выживания. Видно, что чем старше пассажир, тем с больше вероятностью он выживает. Вероятнее всего возраст также будет коррелироваться с классом пассажира, проверим.

```python
ageSurvData = trainData
ageSurvData['IntervalAge'] = ((ageSurvData['Age'] // 10) * 10).dropna()
ageSurvData = ageSurvData[['Survived', 'IntervalAge']].dropna()
ageSurvData['IntervalAge'] = (ageSurvData['IntervalAge'].apply(int)).apply(str) + '-' + (ageSurvData['IntervalAge'].apply(int) + 10).apply(str)
ageSurvData = ageSurvData.groupby(['IntervalAge', 'Survived']).size().reset_index(name='count')
ageSurvData = ageSurvData[ageSurvData['Survived'] == 1].merge(ageSurvData[ageSurvData['Survived'] == 0][['IntervalAge', 'count']], left_on='IntervalAge', right_on='IntervalAge')
ageSurvData = ageSurvData.drop(columns = ['Survived'],axis = 1)
ageSurvData = ageSurvData.rename(columns={"count_x": "countSurv", 'count_y': "countDie"})
ageSurvData['percantageSurv'] = (ageSurvData['countSurv'] / (ageSurvData['countSurv'] + ageSurvData['countDie'])) * 100
ageSurvData['percantageDie'] = (ageSurvData['countDie'] / (ageSurvData['countSurv'] + ageSurvData['countDie'])) * 100
ageSurvData = ageSurvData.drop(columns = ['countSurv', 'countDie'],axis = 1)
fig, ax = plt.subplots()
ageSurvData.plot(kind='bar', ax=ax, stacked=True, x="IntervalAge", xlabel='Возрастной интервал', ylabel='%', figsize=(12,5))
plt.legend(["Выжившие", "Погибшие"])
```

# Анализируем корреляцию возрастного интервала и класса пассажира. Видно, что в первом классе преобладают пассажиры более старшие по возрасту. Также если сравнивать с предыдущим графиком, что видно, что в интервалах 0-10, 20-30 и 30-40 больше всего пассажировов 3 класса и одновременно больше всего погибших в процентном соотношении.

```python
ageClassData = trainData
ageClassData['IntervalAge'] = ((ageClassData['Age'] // 10) * 10).dropna()
ageClassData = ageClassData[['Pclass', 'IntervalAge']].dropna()
ageClassData['IntervalAge'] = (ageClassData['IntervalAge'].apply(int)).apply(str) + '-' + (ageClassData['IntervalAge'].apply(int) + 10).apply(str)
ageClassData = ageClassData.groupby(['IntervalAge', 'Pclass']).size().reset_index(name='count')
ageClassData = ageClassData[ageClassData['Pclass'] == 1].merge( \
                                                               ageClassData[ageClassData['Pclass'] == 2][['IntervalAge', 'count']], left_on='IntervalAge', right_on='IntervalAge').merge( \
                                                               ageClassData[ageClassData['Pclass'] == 3][['IntervalAge', 'count']], left_on='IntervalAge', right_on='IntervalAge')
ageClassData = ageClassData.drop(columns = ['Pclass'],axis = 1)
ageClassData = ageClassData.rename(columns={"count_x": "1 class", 'count_y': "2 class", "count": "3 class"})
ageClassData
fig, ax = plt.subplots()
ageClassData.plot(kind='bar', ax=ax, stacked=True, x="IntervalAge", xlabel='Возрастной интервал', figsize=(12,5))
plt.legend(["1 класс", "2 класс", "3 класс"])
```

# Анализ влияния стоимость билета на шансы выжить. Также наблюдается корреляция между ценой билета и шансами на выживание - чем выше заплатил, тем больше шансов.

```python
dataFare = trainData[['Survived', 'Fare']].dropna()
dataFare['FareCategory'] = pd.cut(dataFare['Fare'], bins=[0,8,15,45,120], labels=['Low','Mid', 'High_Mid','High'])
dataFare = dataFare.drop(columns=['Fare'], axis=1)
dataFare = dataFare.groupby(['FareCategory', 'Survived']).size().reset_index(name='count')
dataFare = dataFare[dataFare['Survived'] == 1].merge(dataFare[dataFare['Survived'] == 0], left_on="FareCategory", right_on="FareCategory")
dataFare = dataFare.drop(columns=['Survived_x', 'Survived_y'], axis=1)
dataFare = dataFare.rename(columns={'count_x': 'countSurv', 'count_y': 'countDie'})
fig, ax = plt.subplots()
dataFare.plot(kind='bar', ax=ax, stacked=True, x="FareCategory", xlabel='Тариф', figsize=(12,5))
plt.legend(["Выжившие", "Погибшие"])
```
